```python
import urllib.request  
import bs4 as bs
import pandas as pd
from pandas import DataFrame

sauce = urllib.request.urlopen('http://openinsider.com/screener?s=&o=&pl=1&ph=10&ll=&lh=&fd=7&fdr=&td=0&tdr=&fdlyl=&fdlyh=&daysago=&xp=1&vl=&vh=&ocl=&och=&sic1=-1&sicl=100&sich=9999&grp=0&nfl=&nfh=&nil=&nih=&nol=&noh=&v2l=&v2h=&oc2l=&oc2h=&sortcol=0&cnt=500&page=1').read()
soup=bs.BeautifulSoup(sauce,'lxml') 
data=[]
temps= pd.DataFrame
list_of_insiderbuys=[]
table = soup.find('table', {'class': 'tinytable'})
table_rows = table.find_all('tr')
for tr in table_rows:
    td = tr.find_all('td')
    row = [i.text for i in td]
    if len(row) == 0:
        continue
    else:
        list_of_insiderbuys.append((row[1:13]))
df= DataFrame(data=list_of_insiderbuys, columns=['Filing Date','Trade Date','Ticker','Company','Insider Name','Insider Title','Trade Type','Share Price','Shares Traded','Shares Owned','Own chg','Value Traded'])
df.to_csv('insiderbuysweek.csv', index=False, encoding='utf-8') 
df
```
